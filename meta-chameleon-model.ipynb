# What exactly is the Chameleon model?

The Chameleon model is a family of models that are capable of understanding and generating text and images. But instead of separate encoders for images and text, the Chameleon family uses a token-based early-diffusion method. Both text and images are converting into a unifom sequence of tokens at the initial processing stage. A single transformer processes these mixed-modal tokens which eliminates the need for separate modality-specific encoders. According to the creators, this method is much more scalable and efficient than previous approaches.

## Step 3: Download the weights

In the cell below, we do download the 7B param weights. Make sure to add your link to the `link` variable

````python
import subprocess

link = "<link>"

command = [
    "python", "-m", "chameleon.download_data", 
    link
]

proc = subprocess.Popen(command, stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

output, errors = proc.communicate(input='7B\n')

print(output)
if errors:
    print("Errors:", errors)
````
